```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Values for k
k_values = [2, 3, 5, 10, 20, 30, 100, 1000]

# Generate a range of x values
x = np.linspace(-8, 200, 1000)

# Plot the Chi-squared PDF for each k, normalized by the degrees of freedom
for k in k_values:
    
    # Normalize by dividing by degrees of freedom
    z = (x / k  - 1) / (2/k)**0.5
    
    # Calculate the PDF of the chi-squared distribution
    y = chi2.pdf(x, k) * (k*(2/k)**0.5)
    
    # Plot the normalized PDF
    plt.plot(z, y, label=f'k={k}')
    
# Add labels and title
plt.xlabel('x')
plt.ylabel(r'$f_k(x)$')
plt.title(r'$\frac{\frac{\chi^2}{k} - 1}{\sqrt{2/k}}$ PDF')

plt.xlim(-4, 4)

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
