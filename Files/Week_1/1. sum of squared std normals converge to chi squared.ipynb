```python
import numpy as np
import matplotlib.pyplot as plt

# Redefine parameters
k_values = [1, 2, 3, 5, 10]  # Selected k values
nsims = 10000  # Number of simulations

# Dictionary to store the sum of squares for each k
sum_of_squares_selected = {k: [] for k in k_values}

# Simulation
for k in k_values:
    for _ in range(nsims):
        # Generate a sample of k standard normal random variables
        sample = np.random.normal(size=k)
        # Calculate the sum of squares and store it
        sum_of_squares_selected[k].append(np.sum(sample**2))

# Plotting histograms on the same chart
plt.figure(figsize=(10, 6))
for k in k_values:
    plt.hist(sum_of_squares_selected[k], bins=30, alpha=0.5, label=f'k={k}', density=True)

plt.title('Histograms of Sum of Squares for Selected k Values')
plt.xlabel('Sum of Squares')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
from scipy.stats import chi2

# Plotting histograms with corresponding chi-squared PDFs
plt.figure(figsize=(10, 6))

for k in k_values:
    # Plot histogram
    plt.hist(sum_of_squares_selected[k], bins=30, alpha=0.5, label=f'k={k} (Histogram)', density=True)

    # Generate x values for chi-squared PDF
    x = np.linspace(min(sum_of_squares_selected[k]), max(sum_of_squares_selected[k]), 1000)
    
    # Plot chi-squared PDF
    plt.plot(x, chi2.pdf(x, k), label=f'k={k} (Chi-squared PDF)', linewidth=2)

plt.title('Histograms and Chi-Squared PDFs for Selected k Values')
plt.xlabel('Sum of Squares')
plt.ylabel('Density')
plt.xlim(0, 20)
plt.ylim(0, 1)

plt.legend()
plt.show()
```
