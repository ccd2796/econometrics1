### 1. From normal distribution to chi squared (with matrices)

A Chi Squared distribution with $n$ degrees of freedom is defined as the sum of the squares of $n$ standard normal uncorrelated random variables

$$ x_i \sim N(0, 1)$$

\begin{equation}
x = \begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{pmatrix} \sim N_n(0, I_n)
\end{equation}

$$Q = \sum_n x_i^2 = x^T x = x^T I_n x = \|x\|^2 \quad \text{and $\; Q \sim \chi^2(n)$} $$

Consider a projection matrix $P$ with $rank(P)=r < n$. From the previous conclusion recall that $P=VV^T$ where $V_{n \times r}$ is the matrix with the eigenvectors of $P$ associated with non null eigenvalues.


Define  $z_{r\times 1} = V^T x_{n \times 1}$. What's its distribution?


$$ z = V^Tx \sim N(0, V^TV) = N_r(0, I_r) $$


Because $z$ comes form a $r$-dimensional standard multivariate normal distribution, it follows that **$z^Tz$ has a $\chi^2(r) $ distribution**


$$ z^Tz = x^TVV^Tx = x^TPx = x^TP^TPx = \|Px\|^2 \sim \chi^2(r) $$

When a vector is projected on an lower dimensional space, some information is lost. If this vector is random, the lower variability imposed on the elements of the projected vector will make its norm likely smaller.

Be careful when trying to find the distribution of the projected vector:

$$ Px \sim N(0, P^TP) = N_n(0, P) $$

This is a degenerate multivariate normal distribution, $P$ is not full rank, therefore the covariance matrix is not positive definite.


***
**Conclusion:**

Consider an $n$-dimensional standard multivariate distribution $x$ and a projection matrix $P$ with $rank(P)=r$.

The distribution of $x^Tx = \|x\|^2 $ is Chi squared with $n$ degrees of freedom.

The distribution of $x^TPx = \|Px\|^2 $ is Chi squared with $r$ degrees of freedom.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2


def gram_schmidt(A):
    """
    Perform the Gram-Schmidt process on the columns of matrix A.

    Parameters:
    - A: a numpy array of shape (n, r) where n is the number of dimensions and r is the number of vectors.

    Returns:
    - Q: a numpy array of shape (n, r) containing the orthonormal basis vectors as its columns.
    """
    n, r = A.shape
    Q = np.zeros((n, r))
    for i in range(r):
        # Start with the original vector
        v_i = A[:, i]
        for j in range(i):
            # Subtract the component of v_i in the direction of the jth orthonormal vector
            v_i -= np.dot(Q[:, j], A[:, i]) * Q[:, j]
        # Normalize v_i to get the ith orthonormal vector
        Q[:, i] = v_i / np.linalg.norm(v_i)
    return Q


# Set dimensions and number of vectors
n = 10  # Dimension
r = 4   # Number of random vectors

# Generate r random vectors within the n-dimensional unit ball
np.random.seed(0)  # For reproducibility
random_vectors = np.random.normal(0, 1, (n, r))
for i in range(r):
    random_vectors[:, i] /= np.linalg.norm(random_vectors[:, i])  # Normalize to lie within the unit ball

# Create matrix V from these vectors
V = gram_schmidt(V)

# Create projection matrix P = V * V.T
P = V @ V.T

# Generate a sample of 1000 vectors from a 10-dimensional multivariate standard normal distribution
sample_vectors = np.random.normal(0, 1, (n, 10000))

# Transform them using the projection matrix P
transformed_vectors = P @ sample_vectors

# Calculate the norm of the vectors, both the sample and the transformed
norms_original = np.linalg.norm(sample_vectors, axis=0)**2
norms_transformed = np.linalg.norm(transformed_vectors, axis=0)**2

# Plot comparing the histograms of the norms
plt.figure(figsize=(8, 5))
plt.hist(norms_original, bins=30, alpha=0.5, label='Random $N_n(0,I)$ vector', density=True)
plt.hist(norms_transformed, bins=30, alpha=0.5, label='Projected vector in $r$=' + str(r), density=True)

# Overlay the PDF of a chi-squared with 10 degrees of freedom and a chi-squared with 4 degrees of freedom
x = np.linspace(0, max(norms_original.max(), norms_transformed.max()), 1000)
plt.plot(x, chi2.pdf(x, df=n), 'r-', label='Chi-squared PDF (10 dof)')
plt.plot(x, chi2.pdf(x, df=r), 'g-', label='Chi-squared PDF (4 dof)')

plt.xlabel('Squared Norm')
plt.ylabel('Density')
plt.title('Comparison of squared norms distribution: ' + 'n=' + str(n) + ' r=' + str(r))
plt.legend()
plt.show()


```
