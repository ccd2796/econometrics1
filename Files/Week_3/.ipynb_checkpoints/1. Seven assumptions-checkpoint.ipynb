### Seven assumptions ###

The seven assumptions will allow to derive certain properties about the estimated models.


**Assumption 1**: fixed regressors. The $n$ observations of the explanatory variables are fixed numbers. Satisfy $ \sum (x_i - \bar{x})^2 $ > 0

**Assumption 2**: The $n$ **disturbances** $\epsilon_i$ are random variables with $E[\epsilon_i] = 0$

**Assumption 3**: Homokedasticity: $E[\epsilon_i^2]= \sigma^2$

**Assumption 4**: $E[\epsilon_i\epsilon_j]= 0$

**Assumption 5**: Parameters $\beta$ and $\sigma > 0$ are fixed unknown numbers. 

**Assumption 6**: Linear model: data was generated by $y = X\beta  + \epsilon$

**Assumption 7**: $\epsilon \sim N_n(\mu, \Sigma)$

Assumption 1 Works when working with controlled experiments. In practice, this helps to handle the expectation operator.

Assumptions 2-4 say that the disturbances follow a distribution with finite first and second moments, and they are uncorrelated. With 7, this becomes $\epsilon \sim N_n(0,\sigma^2 I_n)$
