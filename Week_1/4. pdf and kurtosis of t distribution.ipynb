```python
from scipy.stats import t
import numpy as np
import matplotlib.pyplot as plt


# Define values for degrees of freedom (v)
v_values = [1, 5, 10, 20, 35, 100]

# Generate a range of x values
x = np.linspace(-5, 5, 1000)

# Plotting the PDFs of the t-distribution for each v
plt.figure(figsize=(10, 6))

for v in v_values:
    plt.plot(x, t.pdf(x, df=v), label=f'v={v}')

plt.title('PDFs of t-Distribution for Different Degrees of Freedom')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define a range for degrees of freedom
v_range = np.arange(5, 100)  # Starting from 5 since kurtosis is undefined for v <= 4

# Calculate kurtosis for each degree of freedom
kurtosis_values = [6 / (v - 4) + 3 for v in v_range]

# Plotting kurtosis against degrees of freedom
plt.figure(figsize=(10, 6))
plt.plot(v_range, kurtosis_values, label='Kurtosis of t-Distribution')

plt.title('Kurtosis vs. Degrees of Freedom for t-Distribution')
plt.xlabel('Degrees of Freedom (v)')
plt.ylabel('Kurtosis')
plt.grid(True)
plt.legend()
plt.show()
```
